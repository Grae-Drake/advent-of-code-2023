## Part 1
What would your total score be if everything goes exactly according to your strategy guide?


```python
def get_data(day, test):
    path = 'Input/Day_{}{}.txt'.format(day, "_test" if test else "")
    print("Opening data file at {}".format(path))
    with open(path) as file:
        lines = ''.join(file.readlines()).rstrip()
    return lines
```

```python
# Update each day if necessary based on input format
def process_data(data):
    return [match.split(' ') for match in data.split('\n')]
```

```python
# Confirm data is loaded & processed.

day = 2
test = False

data = get_data(2, test)
data = process_data(data)
data
```

```python
def solve_a(data):
    opponent_moves = "ABC"
    your_moves = "XYZ"
    
    
    def outcome(match):
        opponent_move = opponent_moves.index(match[0])
        your_move = your_moves.index(match[1])
        return ((your_move - opponent_move + 1) % 3) * 3
    
    move_scores = [your_moves.index(match[1]) +1 for match in data]
    outcome_scores = [outcome(match) for match in data]
    
    return sum(move_scores) + sum(outcome_scores)
    
```

```python
solve_a(data)
```

## Part 2
Following the Elf's instructions for the second column, what would your total score be if everything goes exactly according to your strategy guide?

```python
def solve_b(data):
    opponent_moves = "ABC"
    outcomes = "XYZ"
    
    def your_move(match):
        opponent_move = opponent_moves.index(match[0])
        outcome = outcomes.index(match[1])
        return (opponent_move + outcome - 1) % 3 + 1
    
    move_scores = [your_move(match) for match in data]
    outcome_scores = [outcomes.index(match[1]) for match in data] * 3
    
    return sum(move_scores) + sum(outcome_scores)
```

```python
solve_b(data)
```
